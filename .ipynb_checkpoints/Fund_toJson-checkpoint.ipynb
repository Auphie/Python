```python
from bs4 import BeautifulSoup

with open('/home/ubuntu/PythonLesson/PTT/M_1516190862_A_36C.html') as f:
    html = f.read()

soup = BeautifulSoup(html, 'lxml')
```

```python
# 這行裡面剛好就有所有的header資訊

soup.select('.article-meta-value')
```

```python
# 將選出來的list轉成.text,可以將html元素去除

head_untag = [tag.text for tag in soup.select('.article-meta-value')]
head_untag
```

```python
head_untag[0]
```

```python
head_untag[3]
```

```python
page = {}
page['author']   = head_untag[0]
page['board']    = head_untag[1]
page['title']    = head_untag[2]
page['pub_date'] = head_untag[3]
```

```python
# 轉成Json檔了

page
```

```python
page['author']
```

```python
print(soup.select_one('#main-content').text.strip())
```

```python
page['content'] = soup.select_one('#main-content').text.strip()
```
